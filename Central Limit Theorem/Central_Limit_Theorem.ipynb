```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Amrendra\\Documents\\student.csv")
```

```python
df["lpa"].plot(kind='kde')
```

```python
# sample size try 50 ,100, 150, 200
sample = []
for i in range(100):
    sample.append(df["lpa"].sample(200).values.tolist())
```

```python

samples = np.array(sample)
```

```python
sampling = samples.mean(axis=1)
```

```python
sns.kdeplot(sampling)
```

```python
sampling.mean() #sampling mean
```

```python
df["lpa"].mean()  #population mean
```

```python
#confidence Interval 95%
sampling.std()/np.sqrt(200)
```

```python
print('lower limit \t\t upper limit')
sampling.mean() - 2*(sampling.std()/np.sqrt(200)) , sampling.mean() + 2*(sampling.std()/np.sqrt(200))
```
